```python
# ==========================================================
# HEARTGUARD - DATA LOADING & INITIAL VALIDATION MODULE
# ==========================================================

import os
import pandas as pd

# ----------------------------------------------------------
# 1. DEFINE BASE PATHS
# ----------------------------------------------------------

BASE_PATH = os.path.join("..", "data", "raw")
PROCESSED_PATH = os.path.join("..", "data", "processed")

# Ensure processed folder exists
os.makedirs(PROCESSED_PATH, exist_ok=True)

# ----------------------------------------------------------
# 2. GENERIC SAFE LOADER FUNCTION
# ----------------------------------------------------------

def load_dataset(filename, sep=","):
    try:
        path = os.path.join(BASE_PATH, filename)
        df = pd.read_csv(path, sep=sep)
        print(f"\n✅ Loaded: {filename}")
        print(f"Shape: {df.shape}")
        return df
    except FileNotFoundError:
        print(f"❌ File not found: {filename}")
    except Exception as e:
        print(f"⚠ Error loading {filename}: {e}")


# ----------------------------------------------------------
# 3. LOAD ALL DATASETS (Corrected)
# ----------------------------------------------------------

framingham = load_dataset("framingham_raw.csv")
cardiovascular = load_dataset("cardiovascular_raw.csv", sep=';')  # FIXED
heart_attack = load_dataset("heart_attack_raw.csv")
heart_failure = load_dataset("heart_failure_raw.csv")
diabetes_binary = load_dataset("diabetes_binary_raw.csv")
diabetes_multiclass = load_dataset("diabetes_multiclass_raw.csv")


# ----------------------------------------------------------
# 4. STANDARDIZE COLUMN NAMES
# ----------------------------------------------------------

def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("/", "_")
    )
    return df

framingham = clean_column_names(framingham)
cardiovascular = clean_column_names(cardiovascular)
heart_attack = clean_column_names(heart_attack)
heart_failure = clean_column_names(heart_failure)
diabetes_binary = clean_column_names(diabetes_binary)
diabetes_multiclass = clean_column_names(diabetes_multiclass)


# ----------------------------------------------------------
# 5. REMOVE DUPLICATES
# ----------------------------------------------------------

def remove_duplicates(df, name):
    before = df.shape[0]
    df = df.drop_duplicates()
    after = df.shape[0]
    print(f"🧹 {name}: Removed {before - after} duplicate rows")
    return df

framingham = remove_duplicates(framingham, "Framingham")
cardiovascular = remove_duplicates(cardiovascular, "Cardiovascular")
heart_attack = remove_duplicates(heart_attack, "Heart Attack")
heart_failure = remove_duplicates(heart_failure, "Heart Failure")
diabetes_binary = remove_duplicates(diabetes_binary, "Diabetes Binary")
diabetes_multiclass = remove_duplicates(diabetes_multiclass, "Diabetes Multiclass")


# ----------------------------------------------------------
# 6. INSPECT DATASETS
# ----------------------------------------------------------

def inspect_dataset(name, df):
    print("\n" + "="*70)
    print(f"DATASET: {name}")
    print("="*70)

    print("\n🔹 Shape:", df.shape)
    print("\n🔹 Columns:", df.columns.tolist())
    print("\n🔹 Missing Values:\n", df.isnull().sum())
    print("\n🔹 Data Types:\n", df.dtypes)
    print("\n🔹 Sample Data:\n", df.head())

inspect_dataset("Framingham", framingham)
inspect_dataset("Cardiovascular", cardiovascular)
inspect_dataset("Heart Attack", heart_attack)
inspect_dataset("Heart Failure", heart_failure)
inspect_dataset("Diabetes Binary", diabetes_binary)
inspect_dataset("Diabetes Multiclass", diabetes_multiclass)


# ----------------------------------------------------------
# 7. SAVE CLEANED VERSION (NO MODIFICATIONS YET)
# ----------------------------------------------------------

def save_clean_copy(df, filename):
    save_path = os.path.join(PROCESSED_PATH, filename)
    df.to_csv(save_path, index=False)
    print(f"💾 Saved cleaned copy: {filename}")

save_clean_copy(framingham, "framingham_clean_v1.csv")
save_clean_copy(cardiovascular, "cardiovascular_clean_v1.csv")
save_clean_copy(heart_attack, "heart_attack_clean_v1.csv")
save_clean_copy(heart_failure, "heart_failure_clean_v1.csv")
save_clean_copy(diabetes_binary, "diabetes_binary_clean_v1.csv")
save_clean_copy(diabetes_multiclass, "diabetes_multiclass_clean_v1.csv")

print("\n🚀 All datasets loaded, validated, deduplicated & saved successfully!")
```
